# Описание данных

Технологический процесс:
* `Rougher feed` — исходное сырье
* `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
  * `Xanthate`— ксантогенат (промотер, или активатор флотации);
  * `Sulphate` — сульфат (на данном производстве сульфид натрия);
  * `Depressant` — депрессант (силикат натрия).
* `Rougher process` (англ. «грубый процесс») — флотация
* `Rougher tails` — отвальные хвосты
* `Float banks` — флотационная установка
* `Cleaner process` — очистка
* `Rougher Au` — черновой концентрат золота
* `Final Au` — финальный концентрат золота

Параметры этапов:
* `air amount` — объём воздуха
* `fluid levels` — уровень жидкости
* `feed size` — размер гранул сырья
* `feed rate` — скорость подачи

## Модель

### Напишем функцию для вычисления итоговой sMAPE.

```python
Image(url= "https://pictures.s3.yandex.net/resources/smape_1576239058.jpg")
```

```python
def smape(target, predict):
    smape = (1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100
    return smape
```

```python
smape_scorer = make_scorer(smape, greater_is_better = False)
```

```python
Image(url= "https://pictures.s3.yandex.net/resources/_smape_1576239054.jpg")
```

```python
def total_smape(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
```

#### Ленейная регрессия

```python
%%time
model_lr = LinearRegression()
scores_rougher = cross_val_score(model_lr, features_train, target_train['rougher.output.recovery'], scoring=smape_scorer, cv=5) 
scores_final = cross_val_score(model_lr, features_train, target_train['final.output.recovery'], scoring=smape_scorer, cv=5) 
total_smape_lr = total_smape(scores_rougher, scores_final)*(-1)
print('Средняя оценка качества модели:', total_smape_lr.mean())
```

#### Дерево решений

```python
STATE = RandomState(state)
```

```python
%%time
model_dtree_rougher = DecisionTreeRegressor(random_state=STATE)
parametrs = {'max_depth': range (2,20)}

grid_dt_rougher = GridSearchCV(model_dtree_rougher,
                               parametrs, 
                               cv=5, 
                               scoring=smape_scorer)

grid_dt_rougher.fit(features_train, target_train['rougher.output.recovery'])

print('Лучшие параметры:', grid_dt_rougher.best_params_)
print('smape_rougher', abs(grid_dt_rougher.best_score_))
```

```python
%%time
model_dtree_final = DecisionTreeRegressor(random_state=STATE)
grid_dt_final = GridSearchCV(model_dtree_final,
                               parametrs, 
                               cv=5, 
                               scoring=smape_scorer)

grid_dt_final.fit(features_train, target_train['final.output.recovery'])

print('Лучшие параметры:', grid_dt_final.best_params_)
print('smape_final', abs(grid_dt_final.best_score_))
```

```python
total_smape_dt = total_smape(grid_dt_rougher.best_score_, grid_dt_final.best_score_)*(-1)
print('total_smape', total_smape_dt)
```

#### Случайный лес

```python
%%time
model_rf_rougher = RandomForestRegressor(random_state=STATE)

parametrs = {
    'max_depth': range (2,20),
    'n_estimators': range(1, 100, 10)
}

grid_rf_rougher = GridSearchCV(model_rf_rougher, 
                               parametrs, 
                               cv=3, 
                               scoring=smape_scorer,
                               n_jobs=-1,
                              )
grid_rf_rougher.fit(features_train, target_train['rougher.output.recovery'])

print('Лучшие параметры: ', grid_rf_rougher.best_params_)
print('smape_rougher', abs(grid_rf_rougher.best_score_))
```
